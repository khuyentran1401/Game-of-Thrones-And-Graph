# Vertex

```python
# Graph in AdjLists

## The Vertex

class Vertex:
    def __init__(self,key):
        self.id = key
        self.connectedTo = {}
    
    def addNeighbor(self,nbr,weight=0):
        self.connectedTo[nbr] = weight

    def __str__(self):
        return str(self.id) + ' connectedTo: ' + str([x.id for x in self.connectedTo])

    def getConnections(self):
        return self.connectedTo.keys()

    def getId(self):
        return self.id

    def getWeight(self,nbr):
        return self.connectedTo[nbr]
    
```

# Traverse the Nodes with Depth First Search

```python
def traverseAll(g):
    
    visited = []
    connected = 0
    for v in g:
        if v not in visited:
            visited_v = dfs(v)
            visited += visited_v
            connected += 1
            print('\n\nNode {} has {} connected components'.format(v.getId(), len(visited_v)))
            print("\n\n")
    
    print("Total number of connected components is {}".format(connected))
    print(len(visited))
```

```python
def dfs(s):
    stack = [s]
    visited = []
    while stack:
        v = stack.pop()
        if v not in visited:
            visited.append(v)
            print (v.getId(), end = ", ") 
            for w in v.getConnections():
                if w not in visited:
                    stack.append(w)
                
    #print(len(visited))
    return visited
```

# Breadth First Search to Find the Distance Between Characters

```python
def bfs(g,s,t):
    visited = {}
    d = {}
    for v in g:
        visited[v] = False
        d[v] = float('inf')
     
    d[s] = 0
    queue = [s]
    visited[s] = True
    
    while queue:
        v = queue.pop(0)
        
        for w in v.getConnections():
            if visited[w] == False:
                visited[w] = True
                queue.append(w)
                d[w] = d[v] + 1
    print("The distance from {} to {} is {}".format(s.getId(),t.getId(), d[t]))
                
```

```python
s = g.getVertex('Addam-Marbrand')
t = g.getVertex('Emmond')
bfs(g,s,t)
```

```python
s = g.getVertex('Shitmouth')
t = g.getVertex('Trystane-Martell')
bfs(g,s,t)
```
